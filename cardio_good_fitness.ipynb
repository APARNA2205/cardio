```python
# Defining a method to print the percentage of data values in the plots

def bar_perc(plot, feature):                                             # Creating method to print percentages.
    total = len(feature)                                                 # Length of the column.
    for p in ar.patches:                                                 # For loop.
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)        # Percentage of each class of the category.
        x = p.get_x() + p.get_width() / 2 - 0.05                         # Width of the plot.
        y = p.get_y() + p.get_height()                                   # Height of the plot.
        ar.annotate(percentage, (x, y), size = 12)                       # Annotate the percentage.
```
